```python
import pandas as pd
from binance import Client
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
api_key = os.getenv("BNB_API_KEY")
api_secret = os.getenv("BNB_SECRET_KEY")
client = Client(api_key, api_secret)
prices = client.get_all_tickers()
prices
```

```python
pd.DataFrame(prices).to_excel('../database/prices/current_prices_1.xlsx')
```

```python
df_current_prices = pd.DataFrame(prices)
```

```python
current_price_coin = df_current_prices.loc[df_current_prices['symbol'] == ('BTC' + 'USDT'), 'price']
```
