```python
import pandas as pd
```

### concat reports to get the full one

```python
df1 = pd.read_csv('../database/export_binance/nahem/nahem_1.csv')
df1['report'] = 'nahem_1'
df2 = pd.read_csv('../database/export_binance/nahem/nahem_2.csv')
df2['report'] = 'nahem_2'
df3 = pd.read_csv('../database/export_binance/nahem/nahem_3.csv')
df3['report'] = 'nahem_3'
df4 = pd.read_csv('../database/export_binance/nahem/nahem_4.csv')
df4['report'] = 'nahem_4'
df5 = pd.read_csv('../database/export_binance/nahem/nahem_5.csv')
df5['report'] = 'nahem_5'
```

```python
merged = pd.concat([df1, df2, df3, df4, df5], axis=0)
merged.reset_index(drop=True, inplace=True)
merged
```

```python
duplicateRowsDF = merged[merged.duplicated()]
duplicateRowsDF_all = merged[merged.duplicated(subset=merged.columns.difference(['report']))]
df_diff = pd.concat([duplicateRowsDF,duplicateRowsDF_all]).drop_duplicates(keep=False)
merged = merged.drop(df_diff.index, axis=0)
merged.to_excel('../database/export_binance/nahem/nahem_merged.xlsx')
```

```python
merged.set_index('UTC_Time', inplace=True)
```
